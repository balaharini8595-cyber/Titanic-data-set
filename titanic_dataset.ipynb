```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")

print("Dataset loaded successfully!\n")
```

```python
print("DataFrame Preview:")
print(df.head())


age_series = df["Age"]
fare_series = df["Fare"]
survived_series = df["Survived"]

print("\nSeries created: Age, Fare, Survived")
```

```python
print("\n--- HEAD ---")
print(df.head())

print("\n--- INFO ---")
print(df.info())

print("\n--- DESCRIBE ---")
print(df.describe())
```

```python
print("\nMissing values count:")
print(df.isnull().sum())


df["Age"].fillna(df["Age"].mean(), inplace=True)


df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)


df.dropna(subset=["Fare"], inplace=True)

print("\nMissing values handled.")
```

```python
print("\nPassenger using loc:")
print(df.loc[0])

print("\nPassenger using iloc:")
print(df.iloc[0])


selected_cols = df[["Age", "Fare", "Survived"]]
print("\nSelected columns preview:")
print(selected_cols.head())
```

```python
df.rename(columns={"Sex": "Gender"}, inplace=True)

print("\nColumns after renaming:")
print(df.columns)


melted_df = pd.melt(
    df,
    id_vars=["PassengerId"],
    value_vars=["Age", "Fare"],
    var_name="Feature",
    value_name="Value"
)

print("\nMelted DataFrame sample:")
print(melted_df.head())


pivot_df = df.pivot_table(values="Survived", index="Gender", aggfunc="mean")
print("\nPivot Table:")
print(pivot_df)
```

```python
age_above_30 = df[df["Age"] > 30]
fare_above_50 = df[df["Fare"] > 50]
female_survived = df[(df["Gender"] == "female") & (df["Survived"] == 1)]

print("\nPassengers Age > 30:", len(age_above_30))
print("Passengers Fare > 50:", len(fare_above_50))
print("Female passengers who survived:", len(female_survived))


sorted_by_fare = df.sort_values(by="Fare", ascending=False)
sorted_by_age = df.sort_values(by="Age")
sorted_by_survival = df.sort_values(by="Survived", ascending=False)

print("\nSorting completed.")
```

```python
survival_by_gender = df.groupby("Gender")["Survived"].mean()


avg_age_by_class = df.groupby("Pclass")["Age"].mean()


survival_by_embarked = df.groupby("Embarked")["Survived"].sum()

print("\nSurvival rate by Gender:")
print(survival_by_gender)

print("\nAverage Age by Passenger Class:")
print(avg_age_by_class)

print("\nSurvival count by Embarked:")
print(survival_by_embarked)
```

```python
category_df = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Class_Category": ["Upper", "Middle", "Lower"]
})


merged_df = pd.merge(df, category_df, on="Pclass", how="left")


joined_df = df.join(category_df.set_index("Pclass"), on="Pclass")


concat_df = pd.concat([df.head(), df.tail()])

print("\nMerge, Join, and Concat completed.")
```

```python
avg_age_survival = df.groupby("Survived")["Age"].mean()

highest_fare = df["Fare"].max()
lowest_fare = df["Fare"].min()

survival_percentage = df["Survived"].mean() * 100

print("\n--- INSIGHTS ---")
print("Average age (Survived vs Not):")
print(avg_age_survival)

print(f"\nHighest Fare: {highest_fare}")
print(f"Lowest Fare: {lowest_fare}")
print(f"Survival Percentage: {survival_percentage:.2f}%")


print("\nPattern Insight:")
print("Females had higher survival rate compared to males.")
print("Higher class passengers tended to survive more.")
```
